```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv(r"/content/C__Users_Alshuruq Company_Desktop_Saiket_Telco_Customer_Churn_Encoded.csv")

# Display the first 5 rows
df.head()
```

```python
# Count the occurrences of each category in the 'Churn_Yes' column
churn_counts = df['Churn_Yes'].value_counts()

# Plotting the bar plot
plt.figure(figsize=(8, 6))
churn_counts.plot(kind='bar', color=['#4CAF50', '#F44336'])
plt.title('Churn_Yes Balance')
plt.xlabel('Churn_Yes')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

### Due to the imbalance in the target column, it is important to carefully handle the data split to ensure that both the training and testing sets preserve the original class distribution. This approach helps prevent model bias and ensures that the model is evaluated accurately.

```python
df.drop(columns=['customerID'], inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Assuming the target column is 'Churn_Yes' and it's imbalanced
X = df.drop(columns=['Churn_Yes'])
y = df['Churn_Yes']

# Perform stratified split to maintain class distribution
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Save the training and testing sets to CSV files
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)

print("Files saved successfully!")
```
